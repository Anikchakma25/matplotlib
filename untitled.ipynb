```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y, marker='o', linestyle='-', color='b', label='Line Plot')

plt.title('Simple Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 30, 40]
y2 = [15, 18, 28, 33, 38]

plt.plot(x, y1, marker='o', linestyle='-', color='b', label='Line 1')

plt.plot(x, y2, marker='x', linestyle='--', color='r', label='Line 2')

plt.title('Multiple Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.legend()

plt.show()
```

scatter plot

```python
import matplotlib.pyplot as plt
import numpy as np

# র‍্যান্ডম x এবং y ভ্যালু তৈরি করা
x = np.random.rand(50)  # 50টি র‍্যান্ডম x ভ্যালু
y = np.random.rand(50)  # 50টি র‍্যান্ডম y ভ্যালু

# স্ক্যাটার প্লট তৈরি
plt.scatter(x, y, color='g', marker='x', label='Scatter Points')

# শিরোনাম এবং লেবেল যোগ করা
plt.title('Random Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# লেজেন্ড দেখানো
plt.legend()

# গ্রাফ দেখানো
plt.show()
```
